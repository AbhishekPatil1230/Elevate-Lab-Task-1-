```python
df_clean = df.copy()
```

```python
df_clean.columns = df_clean.columns.str.strip().str.lower().str.replace(" ", "_")
df_clean.columns
```

```python
print("Duplicates before:", df_clean.duplicated().sum())
df_clean.drop_duplicates(inplace=True)
print("Duplicates after:", df_clean.duplicated().sum())
```

```python
# Fill missing values
df_clean['director'].fillna("Unknown", inplace=True)
df_clean['cast'].fillna("Unknown", inplace=True)
df_clean['country'].fillna("Unknown", inplace=True)
df_clean['rating'].fillna("Unrated", inplace=True)
df_clean['duration'].fillna("Unknown", inplace=True)

# Handle date_added (convert to datetime & drop missing)
df_clean['date_added'] = pd.to_datetime(df_clean['date_added'], errors="coerce")
df_clean.dropna(subset=['date_added'], inplace=True)
```

```python
# type (movie/tv show) -> lowercase
df_clean['type'] = df_clean['type'].str.lower().str.strip()

# country -> title case (e.g., "United States")
df_clean['country'] = df_clean['country'].apply(
    lambda x: ', '.join([c.strip().title() for c in x.split(',')]) if x != "Unknown" else x
)
```

```python
duration_extract = df_clean['duration'].astype(str).str.extract(r'(?P<number>\d+)\s*(?P<unit>[A-Za-z]+)')
df_clean['duration_number'] = pd.to_numeric(duration_extract['number'], errors='coerce').astype('Int64')
df_clean['duration_unit'] = duration_extract['unit'].str.lower().replace(
    {'seasons': 'season', 'season': 'season', 'min': 'min'}
)
```

```python
df_clean['rating'] = df_clean['rating'].str.upper().str.strip()
df_clean['rating'] = df_clean['rating'].replace({'NOT RATED': 'UNRATED', 'NR': 'UNRATED'})
```

```python
print("Final shape:", df_clean.shape)
print("\nMissing values after cleaning:\n", df_clean.isnull().sum())
df_clean.head()
```

```python
df_clean.to_csv("netflix_titles_cleaned.csv", index=False)
print("Cleaned dataset saved as netflix_titles_cleaned.csv")
```
